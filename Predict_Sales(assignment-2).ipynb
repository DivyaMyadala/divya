```python
#Join these datasets using Store column
ass1 = pd.read_csv("store.csv")
ass2 = pd.read_csv("sales.csv")
data = pd.concat([ass1, ass2],axis=1)
data
```

```python
#Fill missing values using appropriate techniques
#data1=data.fillna({'Store':'0','StoreType':'0','Assortment':'0','CompetitionDistance':'0','CompetitionOpenSinceMonth':'0','CompetitionOpenSinceYear':'0','Promo2':'0','Promo2SinceWeek':'0','Promo2SinceYear':'0','PromoInterval':'0','Store':'0','DayOfWeek':'0','Data':'0','Sales':'0','Customers':'0','Open':'0','Promo':'0','StateHoliday':'0','SchoolHoliday':'0'})
data1 =data.fillna(0)
data1
```

```python
#Extract new columns from date column and ignore the Data column
data1['Date'] = pd.to_datetime(data1['Date'])
data1['year'] = data1['Date'].dt.year
data1['month'] = data1['Date'].dt.month
data1['day'] = data1['Date'].dt.day
data1.drop('Date', inplace=True, axis=1)
data1
```

```python
#Ignore the customer column for this analysis, since we might not know the customer count for future dates.
#Hence this column should not be an input to the model
data1.drop('Customers', inplace=True, axis=1)
data1
```

```python
#Convert non-numerical columns to numerical columns using appropriate methods
data1.info()
```

```python
#Convert non-numerical columns to numerical columns using appropriate methods

data1['StoreType'] = data1['StoreType'].astype(str)
data1['StoreType'] = data1['StoreType'].map(lambda x : ord(x)-96)
data1['Assortment'] = data1['Assortment'].astype(str)
data1['Assortment'] = data1['Assortment'].map(lambda x : ord(x)-96)
data1['StateHoliday'] = data1['StateHoliday'].astype(str)
data1['StateHoliday'] = data1['StateHoliday'].map(lambda x :ord(x)-96)
data1.loc[data1['StateHoliday'] < 0, 'StateHoliday'] = 0
data1.loc[data1['StoreType'] < 0, 'StoreType'] = 0
data1.loc[data1['Assortment'] < 0, 'Assortment'] = 0
```

```python
from datetime import datetime
#data1['PromoInterval'] = promointerval
#filt = data1['PromoInterval']!=0
def month_num(y):
    try:
        x = y.split(',')
        
    except Exception as e:
        try:
            y = int(datetime.strptime(y.lower(), '%b').month)
            return str(y)
        except:
            return str(y)
    
    for i in range(0, len(x)):
        try:
            #print('in try2')
            x[i] = int(datetime.strptime(x[i].lower(), '%b').month)
        
            
            #print(x[i])
        
        except Exception as e:
            pass
            #print('in except2')
            #print(e)
    x = ','.join(str(ele) for ele in x)
    return x
    
replaced = data1['PromoInterval'].replace(to_replace={'3,6,Sept,12':'3,6,9,12'}, inplace = False)
data1['PromoInterval'] = data1['PromoInterval'].map(lambda x : month_num(x))
#data2 = data2.map(lambda x : month_num(x))
replaced = data1['PromoInterval'].replace(to_replace={'3,6,Sept,12':'3,6,9,12'}, inplace = False)
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
enc.fit(data1['PromoInterval'])
data1['PromoInterval'] = enc.transform(data1['PromoInterval'])
data1
```

```python
#Split data in to training (80%) and testing (20%) (use random_state=1)
X =data1.iloc[:,:-1].values
y =data1.iloc[:,-1].values
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 1)
print(X_train.shape,y_test.shape)
```

```python
#Standardize all input columns using StandardScaler function in sklearn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_train
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
#Using relevant influencing columns, build a linear regression model
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X_train,y_train)
```

```python
#Report RMSE, R square value
y_pred = regr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
round(np.sqrt(mean_squared_error(y_test,y_pred)),4)
```

```python
round(r2_score(y_test,y_pred),4)
```
