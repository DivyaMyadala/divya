```python
import pandas as pd
import numpy as np
```

```python
div=pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
```

```python
#Adding Average column to file
div['Average'] = 0
```

```python
#Adding SR column to the file
div['SR'] = 0
```

```python
div
```

```python
#created a dataframe for batsman
batsman_per = div[['batsman','inning','ball','total_runs','batsman_runs','extra_runs','is_wicket']]
```

```python
batsman_per
```

```python
#calculating Average of batsman
batsman_per['Average'] = batsman_per['total_runs']/batsman_per['inning']
```

```python
#calculating the SR values
batsman_per['SR'] = batsman_per['total_runs']/batsman_per['ball']
```

```python
batsman_per
```

```python
#6 batsmans should play more than 50 matches
batsman_matches = batsman_per.groupby(['batsman']).sum()['inning'].reset_index(name = 'inning')
batsman_balls = batsman_per.groupby(['batsman']).sum()['ball'].reset_index(name = 'ball')
batsman_avg = batsman_per.groupby(['batsman']).sum()['Average'].reset_index(name = 'Average')
batsman_sr = batsman_per.groupby(['batsman']).sum()['SR'].reset_index(name = 'SR')
```

```python
best_batsman = [batsman_matches, batsman_balls, batsman_avg, batsman_sr]
best_batsman
best=pd.concat(best_batsman, axis=1).sort_values('SR',ascending=False).head(6)
best1 = best.loc[:,~best.columns.duplicated()].copy()
best1
```

```python
best2=best1.groupby('batsman')
```

```python
list_bes6=best2.groups.keys()
```

```python
list(list_bes6)
```

```python
#Add a economic rate column to the dataframe
div['econ_rate'] = 0
```

```python
#individual performance for each bowler
bowler_per = div[['bowler','inning','ball','over','total_runs','extra_runs','is_wicket','bowling_team','econ_rate']]
```

```python
bowler_per
```

```python
#adding bowler econ rate feature to dataframe
bowler_per['econ_rate'] = bowler_per['total_runs']/bowler_per['ball']
bowler_per
```

```python
bowler_matches = bowler_per.groupby(['bowler']).sum()['inning'].reset_index(name = 'inning')
bowler_overs = bowler_per.groupby(['bowler']).sum()['over'].reset_index(name = 'over')
bowler_wickets = bowler_per.groupby(['bowler']).sum()['is_wicket'].reset_index(name = 'is_wicket')
bowler_econ_rate = bowler_per.groupby(['bowler']).sum()['econ_rate'].reset_index(name = 'econ_rate')
```

```python
best_bowler = [bowler_matches, bowler_overs, bowler_wickets, bowler_econ_rate]
best_bowler
best=pd.concat(best_bowler, axis=1).sort_values('econ_rate',ascending=False).head(6)
best5 = best.loc[:,~best.columns.duplicated()].copy()
best5
```

```python
best=pd.concat(best_bowler, axis=1).sort_values('econ_rate',ascending=False).head(5)
best5 = best.loc[:,~best.columns.duplicated()].copy()
best3=best5.groupby('bowler')
list_best5=best3.groups.keys()
list(list_best5)
```

```python
#final player line-up names in python list
result = pd.concat([best1,best5], axis=1)
#result
```

```python
final_players = result[['batsman','bowler']]
#final_players
b1 = final_players['batsman'].dropna().tolist()
```

```python
#batsman list
final_list = [x for x in b1 if str(x) != 'nan']
final_list
```

```python
#bowler list
b2 = final_players['bowler'].tolist()
final_list2 = [x for x in b2 if str(x) != 'nan']
final_list2
```

```python
#final player lineup names(best 6 batsmen and best 5 bowlers)
final_list.extend(final_list2)
```
