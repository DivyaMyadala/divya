```python
#Split the data into test and train with 80% train size and random state 0
X =df_ass.iloc[:,:-1].values
y =df_ass.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X_train, X_test,y_train,y_test = train_test_split(X,y, train_size = 0.2, random_state = 0)
print(X_train.shape,y_test.shape)
```

```python
#Apply decision tree
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# providing the inputs for the scalling purpose
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
classifer = DecisionTreeClassifier(criterion='entropy', random_state=0)
# providing the training dataset
classifer.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_train_pred=classifer.predict(X_train)
y_test_pred=classifer.predict(X_test)
print(accuracy_score(y_train,y_train_pred),accuracy_score(y_test,y_test_pred))
```

```python
#!pip install seaborn
```

```python
#prune the tree and get the best hyper parameters
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path=classifer.cost_complexity_pruning_path(X_train,y_train)
alphas=path['ccp_alphas']
alphas
```

```python
len(alphas)
```

```python
alphas.min(),alphas.max()
```

```python
accuracy_train, accuracy_test=[],[]

for i in alphas:
    classifer = DecisionTreeClassifier(ccp_alpha=i)
    classifer.fit(X_train,y_train)
    y_train_pred = classifer.predict(X_train)
    y_test_pred = classifer.predict(X_test)
    
    accuracy_train.append(accuracy_score(y_train, y_train_pred))
    accuracy_test.append(accuracy_score(y_test, y_test_pred))

sns.set()
plt.figure(figsize=(40,7))
sns.lineplot(y=accuracy_train, x=alphas, label="Train Accuracy")
sns.lineplot(y=accuracy_test, x=alphas, label="Test Accuracy")
plt.xticks(ticks=np.arange(0.00, 0.25, 0.01))
plt.show()
```

```python
#Obtain the Confusion matrix, Accuracy, F1 score and ROC-AUC and plot the ROC curve
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_test_pred))
```

```python
def plot_roc_curve(fpr, tpr):
    plt.plot(fpr, tpr, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_pred))
df_ass, class_label = make_classification(n_samples=100000, n_classes=2, weights=[1,1], random_state=0)
#Split the data into train and test sub-datasets.
trainX, testX, trainy, testy = train_test_split(df_ass,class_label,train_size=0.8, random_state=0)
#Fit a model on the train data.
model = RandomForestClassifier()
model.fit(trainX, trainy)
#Predict probabilities for the test data.
probs = model.predict_proba(testX)
#Keep Probabilities of the positive class only.
probs = probs[:, 1]
#Compute the AUC Score.
auc = roc_auc_score(testy, probs)
print('AUC: %.2f' % auc)
#Get the ROC Curve.
fpr, tpr, thresholds = roc_curve(testy, probs)
#Plot ROC Curve using our defined function
plot_roc_curve(fpr, tpr)
```
